```python
from sklearn.datasets import load_iris

iris = load_iris()
iris
```

```python
# Deskripsi dari Sampel Dataset

print(iris.DESCR)
```

# **Explanatory dan Response Valriable**

```python
# Explanatory Variables (Feature)

X = iris.data
print("(Jumlah Baris, Kolom):")
X.shape
```

```python
X
```

```python
# Response Variables (Target)

y = iris.target
y.shape
```

# **Visualisasi Data**

```python
import matplotlib.pyplot as plt

X = X[:, :2]  # Hanya plt 2 kolom awal (2 feature pertama)

x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5

plt.scatter(X[:, 0], X[:, 1], c = y)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)

print(f"X train\t: {X_train.shape}")
print(f"X test\t: {X_test.shape}")
print(f"y train\t: {y_train.shape}")
print(f"y test\t: {y_test.shape}")
```
